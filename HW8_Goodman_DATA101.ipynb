```python
import pandas as pd

# Dataset
df = pd.read_csv('forestfires.csv')

# Summary
print(df.describe())
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
print(df.duplicated().sum())

# Group by state and fireyear
grouped = df.groupby(['state', 'fire_year'])

# Grouped data
grouped_sum = grouped.sum()

grouped_sum = grouped_sum.drop(columns='fire_month')

recent_fires = grouped_sum.query('fire_year >= 2010').reset_index()

pivot_table = recent_fires.pivot(index='state', columns='fire_year', values='acres_burned')

# Reset index
grouped_reset = grouped_sum.reset_index()

grouped_reset['decade'] = pd.cut(grouped_reset['fire_year'],
                                 bins=[1989, 1999, 2009, 2019],
                                 labels=['1990s', '2000s', '2010s'])

# Check
print(grouped_reset.head(25))

no_year_df = grouped_reset.drop(columns='fire_year')

final_group = no_year_df.groupby(['state', 'decade']).sum()

# Show first 5 rows
print(final_group.head())
```
