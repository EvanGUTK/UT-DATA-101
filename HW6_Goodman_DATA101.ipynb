```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("cars.csv")

print(df.iloc[15:28])

print(df.describe())

print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])

avg_price_fuel = df.groupby("fueltype")["price"].mean()
print(avg_price_fuel)

plt.figure(figsize=(8, 5))
sns.barplot(x="fueltype", y="price", data=df)
plt.title("Price vs Fuel Type")
plt.show()

avg_price_door = df.groupby("doornumber")["price"].mean()
print(avg_price_door)

#Comparing numbers based on csv
if avg_price_fuel["gas"] > avg_price_fuel["diesel"]:
    print("Gas cars are more expensive on average than diesel cars.")
else:
    print("Diesel cars are more expensive on average than gas cars.")

#Comparing numbers based on csv
if avg_price_door["two"] > avg_price_door["four"]:
    print("Two-door cars are more expensive on average than four-door cars.")
else:
    print("Four-door cars are more expensive on average than two-door cars.")


plt.figure(figsize=(8, 5))
sns.scatterplot(x="horsepower", y="price", hue="fueltype", data=df)
plt.title("Horsepower vs Price by Fuel Type")
plt.show()
```
